```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("All libraries imported successfully!")
```

```python
df = pd.read_csv('../data/american_bankruptcy_dataset.csv')

print("Shape:", df.shape)
print("Columns:",df.columns.tolist())
```

```python
print(df['status_label'].value_counts())
```

```python
print(df['fyear'].value_counts().sort_index())
```

```python
print(df.isnull().sum())
```

```python
df.describe().round(2)
```

```python
import matplotlib.pyplot as plt

alive = df['status_label'].value_counts()['alive']
failed = df['status_label'].value_counts()['failed']

plt.figure(figsize=(6,6))
plt.pie([alive, failed], 
        labels=['Alive', 'Failed'],
        autopct='%1.1f%%',
        colors=['#27AE60', '#E74C3C'])
plt.title('Company Status Distribution')
plt.show()
```

```python
bankruptcies_per_year = df[df['status_label'] == 'failed'].groupby('fyear').size()

plt.figure(figsize=(12,5))
plt.bar(bankruptcies_per_year.index, bankruptcies_per_year.values, color='#E74C3C')
plt.title('Number of Bankruptcies Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Bankruptcies')
plt.xticks(rotation=45)
plt.show()
```
